## 基本使用

```python
# 创建代理
agent = InterfaceGenerationAgent()

# 使用自然语言生成界面
result = agent.process_text_request(
    "Generate silicon on sapphire heterostructure interface with vacuum"
)

print(f"成功生成 {result['results']['structures_generated']} 个界面")
print(f"输出目录: {result['output_dir']}")
```

## 使用Structure对象

```python
# 创建硅结构
si_lattice = Lattice.cubic(5.431)
si_coords = [
    [0, 0, 0], [0.25, 0.25, 0.25], [0.5, 0.5, 0], [0.75, 0.75, 0.25],
    [0.5, 0, 0.5], [0.75, 0.25, 0.75], [0, 0.5, 0.5], [0.25, 0.75, 0.75]
]
si_structure = Structure(si_lattice, ['Si']*8, si_coords)

# 创建蓝宝石结构
sapphire_lattice = Lattice.hexagonal(4.76, 12.99)
sapphire_coords = [
    [0, 0, 0.352], [0, 0, 0.648], [0.333, 0.667, 0.082],
    [0.333, 0.667, 0.418], [0.667, 0.333, 0.748], [0.667, 0.333, 0.252],
    [0.306, 0.694, 0.25], [0.694, 0.306, 0.25], [0, 0, 0.25],
    [0.306, 0.694, 0.75], [0.694, 0.306, 0.75], [0, 0, 0.75]
]
sapphire_elements = ['Al', 'Al', 'Al', 'Al', 'Al', 'Al', 'O', 'O', 'O', 'O', 'O', 'O']
sapphire_structure = Structure(sapphire_lattice, sapphire_elements, sapphire_coords)

print("硅结构信息:")
print(f"化学式: {si_structure.composition}")
print(f"晶系: {si_structure.get_space_group_info()[0]}")
print(f"晶格参数: a={si_structure.lattice.a:.3f} Å")
print()
print("蓝宝石结构信息:")
print(f"化学式: {sapphire_structure.composition}")
print(f"晶系: {sapphire_structure.get_space_group_info()[0]}")
print(f"晶格参数: a={sapphire_structure.lattice.a:.3f} Å, c={sapphire_structure.lattice.c:.3f} Å")
```

```python
# 使用Structure对象生成界面
result = agent.generate_interface_from_structures(
    film_structure=si_structure,
    substrate_structure=sapphire_structure,
    vacuum_type="with_vacuum",
    optimization_level="basic",
    output_dir="notebook_interfaces",
    # 自定义匹配参数
    max_area=150,
    max_length_tol=0.04,
    max_angle_tol=0.04,
    termination_ftol=0.18
)

print("界面生成结果:")
print(f"成功: {result['success']}")
if result['success']:
    print(f"生成界面数: {result['results']['structures_generated']}")
    print(f"输出文件: {result['results']['files_created']}")
else:
    print(f"错误: {result['error']}")
```

## 高级代理使用

```python
# 创建高级代理
advanced_agent = AdvancedInterfaceAgent()

# 生成界面并进行分析
result = advanced_agent.generate_interface_from_structures(
    film_structure=si_structure,
    substrate_structure=sapphire_structure,
    vacuum_type="with_vacuum",
    optimization_level="advanced",
    output_dir="advanced_interfaces",
    analysis_options={
        'energy_analysis': True,
        'structure_analysis': True,
        'visualization': False
    }
)

print("高级分析结果:")
if hasattr(result, 'analysis_results') and result.analysis_results:
    print("结构分析完成")
    if 'structure_metrics' in result.analysis_results:
        for i, metrics in enumerate(result.analysis_results['structure_metrics']):
            print(f"匹配 {i}: 原子数={metrics['num_atoms']}, "
                  f"体积={metrics['lattice_params']['volume']:.1f} Å³")
else:
    print("无分析结果")
```

## 参数调优

```python
# 定义不同的参数组合进行测试
parameter_sets = [
    {
        "name": "严格匹配",
        "max_area": 100,
        "max_length_tol": 0.03,
        "max_angle_tol": 0.03,
        "termination_ftol": 0.15
    },
    {
        "name": "宽松匹配",
        "max_area": 200,
        "max_length_tol": 0.06,
        "max_angle_tol": 0.06,
        "termination_ftol": 0.25
    },
    {
        "name": "中等匹配",
        "max_area": 150,
        "max_length_tol": 0.04,
        "max_angle_tol": 0.04,
        "termination_ftol": 0.20
    }
]

# 测试不同参数
results = []
for params in parameter_sets:
    print(f"测试参数集: {params['name']}")
    
    result = agent.generate_interface_from_structures(
        film_structure=si_structure,
        substrate_structure=sapphire_structure,
        vacuum_type="with_vacuum",
        output_dir=f"param_test_{params['name'].replace(' ', '_')}",
        **{k: v for k, v in params.items() if k != 'name'}
    )
    
    results.append({
        'params': params,
        'success': result['success'],
        'structures_generated': result['results']['structures_generated'] if result['success'] else 0
    })
    
    print(f"  结果: {result['results']['structures_generated'] if result['success'] else 0} 个界面")
    print()

# 总结结果
print("参数调优总结:")
for r in results:
    print(f"{r['params']['name']}: {r['structures_generated']} 个界面")
```

## 批量处理

```python
# 定义多个材料组合
material_combinations = [
    {
        "name": "Si-Ge",
        "film_elements": ['Si']*8,
        "film_lattice": Lattice.cubic(5.431),
        "substrate_elements": ['Ge']*8,
        "substrate_lattice": Lattice.cubic(5.658)
    },
    {
        "name": "GaAs-Si",
        "film_elements": ['Ga', 'Ga', 'Ga', 'Ga', 'As', 'As', 'As', 'As'],
        "film_lattice": Lattice.cubic(5.653),
        "substrate_elements": ['Si']*8,
        "substrate_lattice": Lattice.cubic(5.431)
    }
]

# 批量生成界面
batch_results = []
for combo in material_combinations:
    print(f"处理: {combo['name']}")
    
    # 创建结构 (这里使用简化的坐标)
    film_coords = [[0, 0, 0], [0.25, 0.25, 0.25]] * 4  # 简化的坐标
    substrate_coords = [[0, 0, 0], [0.25, 0.25, 0.25]] * 4
    
    film_struct = Structure(combo['film_lattice'], combo['film_elements'], film_coords)
    substrate_struct = Structure(combo['substrate_lattice'], combo['substrate_elements'], substrate_coords)
    
    result = agent.generate_interface_from_structures(
        film_structure=film_struct,
        substrate_structure=substrate_struct,
        vacuum_type="with_vacuum",
        output_dir=f"batch_{combo['name']}"
    )
    
    batch_results.append({
        'combination': combo['name'],
        'success': result['success'],
        'structures': result['results']['structures_generated'] if result['success'] else 0
    })
    
    print(f"  完成: {result['results']['structures_generated'] if result['success'] else 0} 个界面")
    print()

print("批量处理总结:")
for r in batch_results:
    status = "✓" if r['success'] else "✗"
    print(f"{status} {r['combination']}: {r['structures']} 个界面")
```

## 实用技巧

```python
# 1. 快速参数测试函数
def test_parameters(agent, film_struct, substrate_struct, param_dict):
    """测试不同参数组合"""
    result = agent.generate_interface_from_structures(
        film_structure=film_struct,
        substrate_structure=substrate_struct,
        output_dir=f"test_{param_dict.get('name', 'unknown')}",
        **{k: v for k, v in param_dict.items() if k != 'name'}
    )
    return {
        'params': param_dict,
        'success': result['success'],
        'count': result['results']['structures_generated'] if result['success'] else 0
    }

# 2. 结果分析函数
def analyze_results(results):
    """分析生成结果"""
    successful = [r for r in results if r['success']]
    total_structures = sum(r['count'] for r in successful)
    
    print(f"成功率: {len(successful)}/{len(results)} ({len(successful)/len(results)*100:.1f}%)")
    print(f"总界面数: {total_structures}")
    print(f"平均界面数: {total_structures/len(successful):.1f}" if successful else "平均界面数: 0")
    
    return {
        'success_rate': len(successful)/len(results),
        'total_structures': total_structures,
        'avg_structures': total_structures/len(successful) if successful else 0
    }

# 使用示例
test_params = [
    {'name': 'strict', 'max_area': 80, 'max_length_tol': 0.02},
    {'name': 'relaxed', 'max_area': 200, 'max_length_tol': 0.08}
]

test_results = []
for params in test_params:
    result = test_parameters(agent, si_structure, sapphire_structure, params)
    test_results.append(result)
    print(f"{params['name']}: {result['count']} 个界面")

analysis = analyze_results(test_results)
print(f"\n分析结果: 成功率 {analysis['success_rate']*100:.1f}%")
```
